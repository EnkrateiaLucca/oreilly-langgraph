```python
from openai import OpenAI

client = OpenAI()
```

```python
def llm_call1(prompt: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4.1",
        messages=[{"role": "user", "content": prompt}]
    )
    
    return response.choices[0].message.content


llm_call1("tell me a simple joke")    
```

```python
def llm_call2(prompt: str) -> str:
    response = client.chat.completions.create(
        model="gpt-4o-mini", 
        messages=[{"role": "user","content": prompt}]
    )
    
    return response.choices[0].message.content

llm_call2("hi")
```

```python
def chain_pattern_2_llms(prompt: str) -> str:
    output1 = llm_call1(prompt)
    print("-----OUTPUT 1 ------")
    print(output1)
    input_prompt_2 = f"""
    Take this joke below:
        {output1}
    and break it down into setup and punchline.
    Your output should be structured like this:
    Setup: [write the setup extracted]
    Punchline: [write the punchline extracted]
    """
    output_joke_analysis = llm_call2(input_prompt_2)
    
    print("-----OUTPUT 2 ------")
    print(output_joke_analysis)
    
    return output_joke_analysis
    
chain_pattern_2_llms("Create a simple joke about pandas")
```
